```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def cost(prediction):
    # prediction это результат (выход) нейронной сети 
    return (prediction - 4) ** 2

def deriv_cost(prediction):
    return 2*(prediction-4)

def slope(prediction):
    # same deriv_cost
    return 2*(prediction-4)

def derivative(x, func):
    h = 0.001
    return (func(x+h) - func(x)) / h
```

```python
values = list(range(-20, 21))
val_func = list(map(cost, values))
val_deriv = list(map(deriv_cost, values))

plt.grid(True)
plt.plot(val_func)
plt.plot(val_deriv)
plt.ylabel('f(x)')
plt.xlabel('x')
plt.show()
```

# Approximation

```python
predict = 6
while cost(predict) > 0.1:
    predict = predict - 0.01 * slope(predict)
    print(predict, slope(predict))
```
