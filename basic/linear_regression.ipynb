```python
import matplotlib.pyplot as plt
```

```python
X = [1,2,4]
Y = [2,4,5]

w = 0.25
b = 0.68

def wags(pats):
    return w * pats + b

def cost(w, b):
    err = 0
    for xi, yi in zip(X, Y):
        err += (wags(xi) - yi)**2
        
    return err


plt.plot(X, Y, 'go')
plt.grid(True)
plt.show()
```

<hr>
first prediction

```python
O = [wags(i) for i in X]

plt.plot(X, Y, 'go')
plt.plot(X, O, 'bo')
plt.grid(True)
plt.show()
```

```python
print(cost(w, b))
```

approximation

```python
def dcdw(w, b):
    summ = 0
    for xi, yi in zip(X, Y):
        summ += 2*((w*xi+b) - yi) * xi
    return summ

def dcdb(w, b):
    summ = 0
    for xi, yi in zip(X, Y):
        summ += 2*((w*xi+b) - yi)
    return summ
```

```python
for _ in range(100):
    w = w - 0.05 * dcdw(w, b)
    b = b - 0.05 * dcdb(w, b)
    print(w, b)
    
```

```python
O = [wags(i) for i in X]

plt.plot(X, Y, 'go')
plt.plot(X, O, 'bo')
plt.grid(True)
plt.show()
```
