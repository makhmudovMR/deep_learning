```python
import matplotlib.pyplot as plt
```

```python
def func(x):
    return x**2

def deriv_func(x):
    return x*2

values1 = list(range(-10, 11))

val_func = list(map(func, values1))
val_deriv = list(map(deriv_func, values1))


plt.plot(val_func)
plt.plot(val_deriv)
plt.ylabel('some numbers')
plt.show()
```
